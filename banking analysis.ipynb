```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Banking.csv")
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Med','High']

df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right='False')
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
df.columns
```

```python
categorcal_col = df[['BRId', 'GenderId','Amount of Credit Cards','Nationality', 'Occupation', 'Properties Owned','Risk Weighting','IAId','Income Band']]

for col in categorcal_col:
    print(f"value counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[['BRId', 'GenderId','Amount of Credit Cards','Nationality', 'Occupation', 'Properties Owned','Risk Weighting','IAId','Income Band']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
df.shape
```

```python
print(df.isna().sum())
```

```python
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], format='%d-%m-%Y')
```

```python
categorical_cols = ['Client ID', 'GenderId', 'Nationality', 'Occupation', 'Properties Owned', 'Risk Weighting', 'BRId', 'IAId']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='Nationality', hue='GenderId')
plt.title('Nationality Distribution by Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 30))
sns.countplot(data=df, y='Occupation', order=df['Occupation'].value_counts().index)
plt.title('Occupation Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Bank Deposits'], bins=30, kde=True)
plt.title('Bank Deposits Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Bank Deposits', y='Business Lending', hue='GenderId')
plt.title('Business Lending vs. Bank Deposits')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Properties Owned', y='Bank Deposits', hue='Risk Weighting')
plt.title('Bank Deposits by Properties Owned and Risk Weighting')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Loyalty Classification', order=df['Loyalty Classification'].value_counts().index)
plt.title('Loyalty Classification Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Risk Weighting', hue='GenderId')
plt.title('Risk Weighting by Gender')
plt.show()
```

```python
# Top 5 most common occupations
print(df['Occupation'].value_counts().head(5))

# Average bank deposit by occupation
print(df.groupby('Occupation')['Bank Deposits'].mean().sort_values(ascending=False).head(10))

# Top 10 customers by bank deposits
print(df[['Client ID', 'Bank Deposits']].sort_values(by='Bank Deposits', ascending=False).head(10))
```
